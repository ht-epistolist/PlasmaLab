# Waveform Acquisition from Tektronix DPO 2014B

This notebook connects to a Tektronix DPO 2014B oscilloscope over Ethernet using VXI-11, retrieves active channel waveforms, saves the data to a CSV file, and plots them inline.

```python
import vxi11
import numpy as np
import matplotlib.pyplot as plt

# Enable inline plotting
%matplotlib inline
```

## 1. Connect to Oscilloscope

Specify the IP address of the oscilloscope. In a direct ethernet connection setup, this is typically `10.10.10.2`.

```python
IP_ADDRESS = "10.10.10.2"

print(f"Connecting to oscilloscope at {IP_ADDRESS}...")
try:
    instr = vxi11.Instrument(IP_ADDRESS)
    idn = instr.ask("*IDN?")
    print(f"Connected: {idn.strip()}")
except Exception as e:
    print(f"Connection failed: {e}")
```

## 2. Detect Active Channels

Queries the oscilloscope to see which analog channels (CH1-CH4) are currently turned on.

```python
active_channels = []
for ch in ["CH1", "CH2", "CH3", "CH4"]:
    is_active = instr.ask(f"SELect:{ch}?").strip()
    if is_active == "1":
        active_channels.append(ch)

print(f"Active channels detected: {active_channels}")
```

## 3. Acquire Waveform Data

For each active channel, we:
1. Configure the data source and formatting (16-bit signed big-endian binary).
2. Retrieve vertical scaling (multiplier, offset) and horizontal scaling (sampling interval) parameters.
3. Download the raw binary curve.
4. Parse the IEEE 488.2 block header and reconstruct the physical waveform.

```python
def parse_ieee_block(raw_data):
    """Parses IEEE 488.2 arbitrary block header."""
    if not raw_data.startswith(b'#'):
        raise ValueError("Invalid block header: does not start with '#'")
    num_digits = int(chr(raw_data[1]))
    data_len = int(raw_data[2:2+num_digits].decode('ascii'))
    return raw_data[2+num_digits : 2+num_digits+data_len]

# Set data format parameters
instr.write("DATa:ENCdg RIBinary")  # Signed integer binary format
instr.write("DATa:WIDth 2")        # 16-bit resolution

waveforms = {}
time_axis = None
yunit = "V"
xunit = "s"

for ch in active_channels:
    print(f"Acquiring {ch}...")
    instr.write(f"DATa:SOUrce {ch}")
    
    # Get record length
    record_length = int(instr.ask("HORizontal:RECOrdlength?").strip())
    instr.write("DATa:STARt 1")
    instr.write(f"DATa:STOP {record_length}")
    
    # Get scaling parameters
    xincr = float(instr.ask("WFMPre:XINcr?").strip())
    xzero = float(instr.ask("WFMPre:XZEro?").strip())
    pt_off = float(instr.ask("WFMPre:PT_Off?").strip())
    ymult = float(instr.ask("WFMPre:YMUlt?").strip())
    yoff = float(instr.ask("WFMPre:YOFf?").strip())
    yzero = float(instr.ask("WFMPre:YZEro?").strip())
    yunit = instr.ask("WFMPre:YUNit?").strip().replace('"', '')
    
    # Download curve
    instr.write("CURVe?")
    raw_curve = instr.read_raw()
    
    # Decode curve data
    curve_bytes = parse_ieee_block(raw_curve)
    raw_ints = np.frombuffer(curve_bytes, dtype=">i2")
    
    # Scale data to physical units
    scaled_data = (raw_ints - yoff) * ymult + yzero
    waveforms[ch] = scaled_data
    
    # Create time axis on first channel run
    if time_axis is None:
        time_axis = xzero + xincr * (np.arange(len(raw_ints)) - pt_off)

print("Waveforms acquired successfully!")
```

## 4. Plot Waveforms

Generates a plot of the acquired channels with appropriate time scaling (e.g., ms, µs, ns).

```python
plt.figure(figsize=(12, 6))

# Determine readable time scale units
max_time = np.max(np.abs(time_axis))
time_multiplier = 1.0
plot_time_unit = xunit

if max_time < 1e-6:
    time_multiplier = 1e9
    plot_time_unit = "ns"
elif max_time < 1e-3:
    time_multiplier = 1e6
    plot_time_unit = "µs"
elif max_time < 1.0:
    time_multiplier = 1e3
    plot_time_unit = "ms"

for ch, data in waveforms.items():
    plt.plot(time_axis * time_multiplier, data, label=ch)

plt.title(f"Waveform Acquisition from DPO 2014B ({idn.strip()})")
plt.xlabel(f"Time ({plot_time_unit})")
plt.ylabel(f"Amplitude ({yunit})")
plt.grid(True)
plt.legend()
plt.show()
```

## 5. Export to CSV

Exports the time and voltage arrays to a CSV file.

```python
csv_filename = "waveform_data.csv"
csv_header = f"Time ({xunit})," + ",".join([f"{ch} ({yunit})" for ch in waveforms.keys()])
csv_data = np.column_stack([time_axis] + [waveforms[ch] for ch in waveforms.keys()])
np.savetxt(csv_filename, csv_data, delimiter=",", header=csv_header, comments="")
print(f"Saved data to {csv_filename}")
```

## 6. Disconnect

Close the connection to the oscilloscope.

```python
instr.close()
print("Connection closed.")
```
